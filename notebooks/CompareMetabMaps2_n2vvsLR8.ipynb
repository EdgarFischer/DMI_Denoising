# Wasser

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Daten laden
water         = np.load("MetabMaps/water_LR8.npy")            # (21, 22, 22, 8)
water_n2v     = np.load("MetabMaps/water_n2v.npy")            # (21, 22, 22, 8)
water_sd      = np.load("MetabMaps/water_LR8_sd.npy")         # (21, 22, 22, 8)
water_n2v_sd  = np.load("MetabMaps/water_n2v_sd.npy")         # (21, 22, 22, 8)

# Slice bei z = 10
z = 10
water_slice        = water[z, :, :, :]
water_n2v_slice    = water_n2v[z, :, :, :]
water_sd_slice     = water_sd[z, :, :, :]
water_n2v_sd_slice = water_n2v_sd[z, :, :, :]

# SD-Quotient berechnen
with np.errstate(divide='ignore', invalid='ignore'):
    sd_ratio_slice = np.where(water_sd_slice != 0, water_n2v_sd_slice / water_sd_slice, np.nan)

# Gemeinsame min/max-Werte für beide Methoden
vmin0, vmax0 = np.nanmin(water_slice),        np.nanmax(water_slice)
vmin1, vmax1 = np.nanmin(water_n2v_slice),    np.nanmax(water_n2v_slice)

# Feste Farbskala für SD-Quotient
vmin2, vmax2 = 0, 1.5

# Layout: 3 Spalten + Colorbar rechts
fig = plt.figure(figsize=(15, 24))
gs = gridspec.GridSpec(8, 4, width_ratios=[1, 1, 1, 0.05], wspace=0.3)

ims2 = []  # Für SD-Quotient

for i in range(8):
    ax0 = plt.subplot(gs[i, 0])
    ax1 = plt.subplot(gs[i, 1])
    ax2 = plt.subplot(gs[i, 2])

    # N2V
    ax0.imshow(water_slice[..., i], cmap='viridis', vmin=vmin0, vmax=vmax0)
    ax0.set_title(f'water LR8, T = {i+1}')
    ax0.axis('off')

    # LR8
    ax1.imshow(water_n2v_slice[..., i], cmap='viridis', vmin=vmin1, vmax=vmax1)
    ax1.set_title(f'water n2v, T = {i+1}')
    ax1.axis('off')

    # SD-Quotient
    im2 = ax2.imshow(sd_ratio_slice[..., i], cmap='plasma', vmin=vmin2, vmax=vmax2)
    ax2.set_title(f'SD(LR8) / SD(N2V), T = {i+1}')
    ax2.axis('off')
    ims2.append(im2)

# Colorbar für SD-Quotient
cax2 = fig.add_subplot(gs[:, 3])
cb2 = fig.colorbar(ims2[0], cax=cax2)
cb2.set_label('SD(N2V) / SD(LR8)')

plt.tight_layout(rect=[0, 0, 0.95, 1])
plt.show()



```

# Glc

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

metabolite = 'Glc'

# Daten laden: erst N2V, dann LR8
n2v        = np.load(f"MetabMaps/{metabolite}_n2v.npy")       # (21, 22, 22, 8)
lr8        = np.load(f"MetabMaps/{metabolite}_LR8.npy")       # (21, 22, 22, 8)
n2v_sd     = np.load(f"MetabMaps/{metabolite}_n2v_sd.npy")    # (21, 22, 22, 8)
lr8_sd     = np.load(f"MetabMaps/{metabolite}_LR8_sd.npy")    # (21, 22, 22, 8)

# Slice bei z = 10
z = 10
slice_n2v   = n2v[z, :, :, :]
slice_lr8   = lr8[z, :, :, :]
slice_n2v_sd = n2v_sd[z, :, :, :]
slice_lr8_sd = lr8_sd[z, :, :, :]

# SD-Quotient berechnen: SD(LR8) / SD(N2V)
with np.errstate(divide='ignore', invalid='ignore'):
    sd_ratio = np.where(slice_n2v_sd != 0, slice_n2v_sd/slice_lr8_sd, np.nan)

# gemeinsame min/max für die Ebenen
vmin0, vmax0 = np.nanmin(slice_n2v), np.nanmax(slice_n2v)
vmin1, vmax1 = np.nanmin(slice_lr8), np.nanmax(slice_lr8)

# feste Farbskala für SD-Quotient
vmin2, vmax2 = 0, 1.5

# Layout: 3 Spalten + Colorbar rechts
fig = plt.figure(figsize=(15, 24))
gs = gridspec.GridSpec(8, 4, width_ratios=[1,1,1,0.05], wspace=0.3)
ims = []

for t in range(8):
    ax0 = plt.subplot(gs[t, 0])
    ax1 = plt.subplot(gs[t, 1])
    ax2 = plt.subplot(gs[t, 2])

    # Spalte 1: N2V
    ax0.imshow(slice_n2v[..., t], cmap='viridis', vmin=vmin0, vmax=vmax0)
    ax0.set_title(f'{metabolite} N2V, T = {t+1}')
    ax0.axis('off')

    # Spalte 2: LR8
    ax1.imshow(slice_lr8[..., t], cmap='viridis', vmin=vmin1, vmax=vmax1)
    ax1.set_title(f'{metabolite} LR8, T = {t+1}')
    ax1.axis('off')

    # Spalte 3: SD-Quotient
    im = ax2.imshow(sd_ratio[..., t], cmap='plasma', vmin=vmin2, vmax=vmax2)
    ax2.set_title(f'SD(N2V) / SD(LR8), T = {t+1}')
    ax2.axis('off')
    ims.append(im)

# Colorbar für SD-Quotient
cax = fig.add_subplot(gs[:, 3])
cb = fig.colorbar(ims[0], cax=cax)
cb.set_label('SD(LR8) / SD(N2V)')

plt.tight_layout(rect=[0, 0, 0.95, 1])
plt.show()

```
