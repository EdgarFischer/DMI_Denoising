```python
# Compare LCModel SNR with IQR

import nibabel as nib
import numpy as np

def load_snr_stats(path, mask=None):
    img = nib.load(path)
    data = img.get_fdata()

    snr = data.reshape(-1)

    # basic cleanup
    snr = snr[np.isfinite(snr)]
    snr = snr[snr > 0]   # LCModel-SNR sollte positiv sein

    if mask is not None:
        mask = mask.reshape(-1).astype(bool)
        snr = snr[mask[:snr.size]]

    mean = np.mean(snr)
    std  = np.std(snr)
    med  = np.median(snr)
    q25, q75 = np.percentile(snr, [25, 75])

    return mean, std, med, q25, q75, snr.size


# --- load both ---
mean_sim, std_sim, med_sim, q25_sim, q75_sim, n_sim = load_snr_stats("SNRLesion.nii")
mean_inv, std_inv, med_inv, q25_inv, q75_inv, n_inv = load_snr_stats("SNR_P08.nii")

print("Simulation:")
print(f"  mean ± SD      = {mean_sim:.2f} ± {std_sim:.2f}")
print(f"  median (IQR)   = {med_sim:.2f} ({q25_sim:.2f} – {q75_sim:.2f})")
print(f"  voxels         = {n_sim}")

print("\nIn vivo (P08):")
print(f"  mean ± SD      = {mean_inv:.2f} ± {std_inv:.2f}")
print(f"  median (IQR)   = {med_inv:.2f} ({q25_inv:.2f} – {q75_inv:.2f})")
print(f"  voxels         = {n_inv}")

```
